```python
import pandas as pd

# Assuming 'data' is your DataFrame
# Select only numeric columns
numeric_cols = data.select_dtypes(include=['number']).columns

# Group by 'Generation', calculate mean for numeric columns, sort by 'Attack' in ascending order, and get the top 10
top_10 = data.groupby('Generation')[numeric_cols].mean().sort_values('Attack', ascending=True).head(10)

print(top_10)
```

```python
data.groupby(['Generation'])[numeric_cols].max().sort_values(['Attack','Defense'], ascending=False).head(10)
```

```python
df = data.groupby(['Generation','Type 1','Legendary']).count()
df.head(5)
```
